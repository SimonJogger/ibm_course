# Capstone project: The Battle of Neighborhoods

## Backgound

I live in Australia thus I want to perform a local case study using Foursquare location data.

Sydney and Melbourne are well-known to be the most famous cities in Australia, and these two cities are both proud of their cultural diversity. However, Sydney and Melbourne are competing with each other in many aspects to show "this city" is the effective cultural capital of Australia.

Hence, in this capstone project, I would like to provide a little evidence that between Sydney and Melbourne, which is the city has a more diverse dining culture.

## Data description

To perform the case study I introduced in the background section, I will choose the central parts of the two cities as the reference locations. Hence, I will choose to use 'Sydney Townhall' as the central point in Sydney, and use 'Melbourne Townhall' as the central point in Melbourne. The exploring radius will be set as 2.5km which is likely to cover the most active areas of the two cities.

The data will be used include:

1. Restaurant name;
2. Restaurant category;
3. Restaurant expense;
4. The number of tips for each restaurant.

Then I will:

1. Use the Foursqure agent to explore the restaurants around the two central points, will a limit of 200;
2. Compare how diverse these restaurants are in terms of styles (say Italian, French, Chinese...);
3. Compare how diverse these restaurants are in terms of expense level (say street food, cheap, medium, expensive, luxury);
4. Compare how popular these restaurants are in terms of the tips in the Foursquare location data.

# Search for Sydney restaurants

## Get the location of the Sydney Town Hall

```python
address = '483 George Street, Sydney, NSW, AU'
geolocator = Nominatim(user_agent='google')
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

## Search for results

```python
search_query = 'restaurant'
radius = 2500
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
```

```python
results = requests.get(url).json()
# extract venue info and put into df
venues =  results['response']['venues']
SYD_dataframe = json_normalize(venues)
#print(SYD_dataframe.columns)
```

```python
SYD_dataframe['Cat'] = SYD_dataframe['categories'].apply(lambda x: x[0]['name'])
```

## Map the restaurants

```python
Sydney_map = folium.Map(location=[latitude, longitude], zoom_start=14) # generate map centred around the Conrad Hotel

# add a red circle marker to represent the Conrad Hotel
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='navy',
    tooltip = 'Sydney Town Hall',
    #popup='1 King William Street',
    fill = True,
    fill_color = 'navy',
    fill_opacity = 1
).add_to(Sydney_map)

# add the Italian restaurants as blue circle markers
for lat, lng, label, cat in zip(SYD_dataframe['location.lat'], SYD_dataframe['location.lng'], SYD_dataframe['name'], SYD_dataframe['Cat']):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color=SYD_color_dict[SYD_style_dict[cat]],
        tooltip=SYD_style_dict[cat],
        popup = label,
        fill = True,
        fill_color=SYD_color_dict[SYD_style_dict[cat]],
        fill_opacity=1
    ).add_to(Sydney_map)

# display map
Sydney_map
```

# Discussion

1. Restaurant styles

Sydney and Melbourne have different restaurant styles in the city certral area. Although Chinese restaurant holds a biggest proportion in both areas, there are more Australian style restaurants in Melbourne than in Sydney. Moreover, there are more Italian restaurants in the central Sydney while more Indian restaurants in Melbourne.

2. Restaurant price level

In both cities central areas, 'moderate' restaurants account for around 50%. However, there are a little bit more 'very expensive' restaurants and a little bit less 'cheap' restaurants in central Sydney than in central Melbourne. This suggests that the typical meal/dinner expensive may be relatively high in central Sydney.

3. Restaurant popularity measured in the number of tips

While the most commented restaurants in Sydney and Melbourne have similar amount of tips, the average number of Sydney restaurant tips is around 30% higher than that in Melbourne (19.76 vs 15.38). This suggests that if measured by Foursqaure data users, the restaurants in Sydney maybe more popular than those in Melbourne.
