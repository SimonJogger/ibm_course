# Search for Sydney restaurants

## Get the location of the Sydney Town Hall

```python
# Get the locational info using google services
address = '483 George Street, Sydney, NSW, AU'
geolocator = Nominatim(user_agent='google')
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

## Search for results

```python
# Form the search query
search_query = 'restaurant'
radius = 2500
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
```

```python
# Get the search results
results = requests.get(url).json()
# extract venue info and put into df
venues =  results['response']['venues']
SYD_dataframe = json_normalize(venues)
#print(SYD_dataframe.columns)
# Reform the category column
SYD_dataframe['Cat'] = SYD_dataframe['categories'].apply(lambda x: x[0]['name'])
```

```python
# Check the results as a dataframe
SYD_dataframe
```

Check the restaurants' categories.

```python
# Check the restaurants' categories.
SYD_dataframe['Cat'].unique()
```

As we can see, some categories are overlapped, thus I set up a dict to organize them by checking both the Cat and Name.

```python
#Set a color dict
SYD_style_dict = {'Chinese Restaurant':'Chinese',
              'Dim Sum Restaurant':'Chinese',
              'Korean Restaurant' : 'Korean',
              'Szechuan Restaurant': 'Chinese',
              'Noodle House':'Chinese',
              'Australian Restaurant':'Australian',
              'Seafood Restaurant': 'Australian',
              'Malay Restaurant':'Malay',
              'Lebanese Restaurant':'Lebanese',
              'Indonesian Restaurant':'Indonesian',
              'Thai Restaurant': 'Thai',
              'Cantonese Restaurant':'Chinese',
              'Art Museum':'Museum',
              'Japanese Restaurant': 'Japanese',
              'Italian Restaurant': 'Italian',
              'Asian Restaurant':'Chinese',
              'Vietnamese Restaurant':'Vietnamese',
              'Restaurant':'Australian',
              'Greek Restaurant':'Greek',
              'Falafel Restaurant':'Falafel',
              'French Restaurant':'French',
              'Bar':'Australian',
              'American Restaurant': 'American'}
```

```python
SYD_color_dict = {'Chinese':'red','Australian':'blue','Malay':'green','Lebanese':'yellow',
                  'Indonesian':'pink','Thai':'orange','Museum':'black','Japanese':'white',
                  'Italian':'teal','Vietnamese':'grey','Greek':'silver','Falafel':'brown',
                  'French':'plum','Korean':'olive', 'American':'blue'}
```

```python
# Create a new Category column after reorganize the restaurants categories
SYD_dataframe['RCat'] = SYD_dataframe['Cat'].apply(lambda x: SYD_style_dict[x])
```

## Map the restaurants

This section maps the obtained restaurants in a Folium map. Each color corresponds to a restaurant category.

Clearly, there are many Chinese restaurants and they are located closely at the central area.

```python
Sydney_map = folium.Map(location=[latitude, longitude], zoom_start=14)

# add a circle marker to represent the Sydney Town Hall
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='navy',
    tooltip = 'Sydney Town Hall',
    fill = True,
    fill_color = 'navy',
    fill_opacity = 1
).add_to(Sydney_map)

# add the different restaurants as colored circles
for lat, lng, label, cat in zip(SYD_dataframe['location.lat'], SYD_dataframe['location.lng'], SYD_dataframe['name'], SYD_dataframe['Cat']):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color=SYD_color_dict[SYD_style_dict[cat]],
        tooltip=SYD_style_dict[cat],
        popup = label,
        fill = True,
        fill_color=SYD_color_dict[SYD_style_dict[cat]],
        fill_opacity=1
    ).add_to(Sydney_map)

# display map
Sydney_map
```

## Explore the restaurants to get the price info and number of tips

```python
# Create a summary df to contain the restaurants' details in Sydney
SYD_venue_summary = pd.DataFrame(columns=['Name','Category','Price','Tips'])

# Use the restaurant id as the index to locate each restaurant
for restaurant_id in SYD_dataframe['id']:
    # Obtain the restaurant name
    venue_name = SYD_dataframe[SYD_dataframe['id']==restaurant_id]['name'].values[0]
    # Obatin the restaurant category
    venue_cat = SYD_dataframe[SYD_dataframe['id']==restaurant_id]['RCat'].values[0]
    # Form a new search query to explore the restaurant detail
    url = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(restaurant_id, CLIENT_ID, CLIENT_SECRET, VERSION)
    # get the query result for a restaurant
    results = requests.get(url).json()
    info =  results['response']['venue']
    info_df = json_normalize(info)
    # Some restaurants do not have pring info, thus I use a try and except structure here
    try:
        price = info_df['price.message'].values[0]
    except:
        price = -1
    # Obtain the number of tips for a restaurant
    tips = info_df['tips.count'].values[0]
    # Put the restaurant info into a df row
    new_df = pd.DataFrame(columns=['Name','Category','Price','Tips'],
                              data=[[venue_name,venue_cat,price,tips]])
    # Append restaurant rows to the summary df
    SYD_venue_summary = SYD_venue_summary.append(new_df)
```
